```python
#How do we remove rows that failed the attention check?
#Dishant Sharma 12530084
import pandas as pd, numpy as np
df = pd.read_csv(r"/MRF_balance.csv")
df = df[df['tad'] == 1].copy()
df.reset_index(drop=True, inplace=True)
print(df.shape)
```

```python
#How do we select only the variables we need?
#Dishant Sharma 12530084
import pandas as pd, numpy as np
df = pd.read_csv(r"/MRF_balance.csv")
VARS = ['equitycapital', 'totalliabilities','totalassets']
df = df[VARS].copy()
print(df.head())
```

```python
#How do we delete rows where more than 50% of values are missing?
#Dishant Sharma 12530084
import pandas as pd, numpy as np
df = pd.read_csv(r"/MRF_balance.csv")
thresh = int(0.5 * df.shape[1])
df = df.dropna(thresh=thresh).copy()
df.reset_index(drop=True, inplace=True)
print(df.shape)
```

```python
#How do we fill missing numeric values with the column mean?
#Dishant Sharma 12530084
import pandas as pd, numpy as np
df = pd.read_csv(r"/MRF_balance.csv")
num = df.select_dtypes(include=[np.number]).columns
for col in num:
 df[col].fillna(df[col].mean(), inplace=True)
print(df.isnull().sum())
```

```python
#Dishant Sharma 12530084
#How do we ensure all numeric columns have the correct data type?
import pandas as pd, numpy as np
df = pd.read_csv(r"/MRF_balance.csv")
num_cols = ['equitycapital', 'totalliabilities','totalassets']
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
print(df.dtypes)
```

```python
#How do we remove duplicates and standardise column names?
#Dishant Sharma 12530084
import pandas as pd, numpy as np
df = pd.read_csv(r"/MRF_balance.csv")
df.drop_duplicates(inplace=True)
df.columns = (df.columns.str.strip().str.lower() .str.replace(r'[^a-z0-9]+','_',regex=True))
df.reset_index(drop=True, inplace=True)
print(df.columns.tolist())
```

```python
#How do we get a summary of the dataset’s structure?
#Dishant Sharma 12530084
import pandas as pd, numpy as np
df = pd.read_csv(r"/MRF_balance.csv")
print(df.shape)
df.info()
print(df.describe())
```

```python
#How do we compute mean, median, min, max and std deviation?
#Dishant Sharma 12530084
import pandas as pd, numpy as np
df = pd.read_csv(r"/MRF_balance.csv")
num = df.select_dtypes(include=[np.number]).columns
stats = pd.DataFrame({
    'Mean':df[num].mean(),  'Median':df[num].median(),
    'Min':df[num].min(),    'Max':df[num].max(),
    'Std':df[num].std()})
print(stats.round(2))
```

```python
#How do we create new derived columns from existing ones?
#Dishant Sharma 12530084
import pandas as pd, numpy as np
df = pd.read_csv(r"/MRF_balance.csv")
df['debt_to_equity'] = df['deposits'] / df['equitycapital']
df['asset_util']     = df['totalliabilities'] / df['totalassets']
df['invest_ratio']   = df['investments'] / df['totalassets']
print(df[['debt_to_equity','asset_util','invest_ratio']].round(3))
```

```python
#How do we filter rows based on a condition?
#Dishant Sharma 12530084
import pandas as pd, numpy as np
df = pd.read_csv(r"/MRF_balance.csv")
high = df[df['totalassets'] > 200000]
print(high[['year','totalassets']])
both = df[(df['totalassets'] > 200000) & (df['tad'] == 1)]
print(both[['year','totalassets','tad']])
```

```python
#How do we compute group-wise statistics using groupby?
#Dishant Sharma 12530084
import pandas as pd, numpy as np
df = pd.read_csv(r"/MRF_balance.csv")
grp = df.groupby('tad')[['equitycapital','totalliabilities','totalassets']].mean().round(2)
print(grp)
```
